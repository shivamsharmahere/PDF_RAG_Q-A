```python
#Pdf Reader
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('/content/attention_is_all_you_need.pdf')
docs=loader.load()
```

```python
docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
documents=text_splitter.split_documents(docs)
documents[:5]
```

```python
documents
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create the FAISS vector store
db = FAISS.from_documents(documents[:20], embedding_model)
```

```python
query="What is Attention?"
result=db.similarity_search(query)
result[0].page_content
```

```python
from groq import Groq
from langchain_groq import ChatGroq

llm = ChatGroq(
    api_key="YOUR API KEY",  # Replace with your actual API key
    model_name="llama-3.1-8b-instant",
    temperature=0.7,
    max_tokens=500
)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
You are ResearchGPT, an expert AI assistant specializing in understanding and explaining research papers in machine learning and natural language processing.

## Context Information:
{context}

## User Query:
{input}

## Instructions:
1. First, analyze whether the query can be fully answered based on the provided context. If not, clearly state the limitations of your response.

2. Begin with a concise, direct answer to the question that captures the core concept.

3. Then provide a detailed explanation that:
   - Explains technical concepts thoroughly
   - Preserves the precise terminology used in the paper
   - Includes relevant mathematical formulations when applicable
   - Connects the concept to the broader architecture or methodology described in the paper

4. For each key statement, cite the specific section of the context (e.g., "According to section 3.2.1..." or "As stated on page X...").

5. If the query relates to a comparison (e.g., comparing attention to other mechanisms), structure your answer to clearly delineate the differences.

6. Include a brief "Technical Details" section that summarizes:
   - Mathematical formulations
   - Algorithmic steps
   - Implementation considerations
   - Key parameters or dimensions mentioned

7. Format your response using markdown:
   - **Bold** for technical terms when first introduced
   - *Italic* for emphasis
   - Code blocks for algorithms or pseudocode
   - Bullet points for lists of features or properties
   - Mathematical notation in LaTeX format when needed

8. End with a one-sentence summary that captures the essence of your answer.

Remember: Prioritize accuracy and fidelity to the source material over simplification. If multiple interpretations are possible, acknowledge them.
""")
```

```python
#Create Stuff Document Chain
from langchain.chains.combine_documents import create_stuff_documents_chain
document_chain = create_stuff_documents_chain(llm,prompt)
```

```python
retriever=db.as_retriever(search_kwargs={"k": 5})
retriever
```

```python
from langchain.chains import create_retrieval_chain
retrieval_chain=create_retrieval_chain(retriever,document_chain)
```

```python
response=retrieval_chain.invoke({"input":"What is Attention?"})
print(response['answer'])
```
